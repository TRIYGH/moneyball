```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
headers = ['playerID','year','AB','H','BB','HBP','SF']
all_batting_data = pd.read_csv('Batting.csv', names=headers, usecols=[0,1,6,8,15,18,20], skiprows=62238)  
```

```python
all_batting_data.head(111)
all_batting_data = all_batting_data.replace('NaN', 0)
all_batting_data.head(111)

```

```python
all_batting_data_1985 = all_batting_data[all_batting_data['year']==1985]
all_batting_data_1985
```

```python
ABD85_bigABs = all_batting_data_1985[all_batting_data_1985['AB']>100]
ABD85_bigABs
```

```python
ABD85_bigABs_sorted = ABD85_bigABs.sort_values('AB',ascending=False)
ABD85_bigABs_sorted
```

```python
OBP = (ABD85_bigABs_sorted['H']+ABD85_bigABs_sorted['BB']+ABD85_bigABs_sorted['HBP'])/(ABD85_bigABs_sorted['AB']+ABD85_bigABs_sorted['BB']+ABD85_bigABs_sorted['HBP']+ABD85_bigABs_sorted['SF'])
OBP
```

```python
# print(ABD85_bigABs_sorted.query(OBP.index[0]))

ABD85_bigABs_sorted['OBP']=OBP
ABD85_bigABs_sorted
```

```python
all_salaries_data = pd.read_csv('Salaries.csv', header=0, usecols=[0,3,4])
```

```python
all_salaries_data_1985 = all_salaries_data[all_salaries_data['yearID']==1985]
all_salaries_data_1985
```

```python
ASD = all_salaries_data_1985
ASD = ASD.drop('yearID', 1)
ASD
```

```python
# ABD85_bigABs_sorted['Salary']=ASD
bigLIST = ABD85_bigABs_sorted.merge(ASD)
bigLIST
```

```python
column_names=['yearID','teamID','lgID','playerID','G_all','GS','G_batting','G_defense','G_p','G_c','G_1b','G_2b','G_3b','G_ss','G_lf','G_cf','G_rf','G_of','G_dh','G_ph','G_pr']
position_data = pd.read_csv('Appearances.csv', names=column_names, skiprows=62206)
position_data
```

```python
position_data85 = position_data[position_data['yearID']==1985]
position_data85
```

```python
position_data85 = position_data85.drop('yearID', 1)
position_data85 = position_data85.drop('teamID', 1)
position_data85 = position_data85.drop('lgID', 1)
position_data85
```

```python
ds_with_sal = bigLIST.merge(position_data85)
```

```python
ds_with_sal
```

```python
# this list will provide one to go thru and look for 'cheap', high OBP players for each position.
#     Salaries in today's dollars could be updated each by 1.0264^30 ( 30 years = 2015 ) 2.64% avg. inflation.
```

```python
ds_with_sal['salary'] = ds_with_sal['salary'].apply(lambda x: x*(1.0264**30))
ds_with_sal

# ds_with_sal_INFL = ds_with_sal['salary'].apply(lambda x: x*(1.0264**30))
# ds_with_sal['salary'].replace(ds_with_sal_INFL)
```
