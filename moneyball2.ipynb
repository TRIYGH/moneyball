```python
column_names=['yearID','teamID','lgID','playerID','G_all','GS','G_batting','G_defense','G_p','G_c','G_1b','G_2b','G_3b','G_ss','G_lf','G_cf','G_rf','G_of','G_dh','G_ph','G_pr']
position_data = pd.read_csv('Appearances.csv', names=column_names, skiprows=62206)
position_data
```

```python
position_data85 = position_data[position_data['yearID']==1985]
position_data85
```

```python
position_data85 = position_data85.drop('yearID', 1)
position_data85 = position_data85.drop('teamID', 1)
position_data85 = position_data85.drop('lgID', 1)
position_data85
```

```python
ds_with_sal = bigLIST.merge(position_data85)
```

```python
ds_with_sal
```

```python
# this list will provide one to go thru and look for 'cheap', high OBP players for each position.
#     Salaries in today's dollars could be updated each by 1.0264^30 ( 30 years = 2015 ) 2.64% avg. inflation.
```

```python
ds_with_sal['salary'] = ds_with_sal['salary'].apply(lambda x: x*(1.0264**30))
save_for_later = ds_with_sal
ds_with_sal
# ds_with_sal_INFL = ds_with_sal['salary'].apply(lambda x: x*(1.0264**30))
# ds_with_sal['salary'].replace(ds_with_sal_INFL)
```

```python
top_picks = ds_with_sal['OBP'] / ds_with_sal['salary'] * 100000
ds_with_sal['TP'] = top_picks
ds_with_sal = ds_with_sal.sort_values('TP',ascending=False)
ds_with_sal
```

```python
ds_with_sal = ds_with_sal[ds_with_sal['AB']>400]
ds_with_sal
```

```python
# input()
order_of_picks = ['G_ss','G_c', 'G_2b', 'G_3b', 'G_1b', 'G_lf', 'G_cf', 'G_rf']
player = []
min_ABs = 100

for each in order_of_picks:
    for index, row in ds_with_sal.iterrows():
        if row[each] > min_ABs:
            player.append(row['playerID'])
            break
    
for i, each in enumerate(order_of_picks):
    print(each, "  --->  ", player[i])
```
